```python
import pandas as pd

#Imports the Pandas Library.
```

```python
eceboard = pd.read_csv('board2.csv')
eceboard

#Loads the 'board2.csv file'
```

1.) Create the following data frames based on the format provided:

<U>Name the file 'Instru' and display the columns with the conditions [“Name”, “GEAS”, “Electronics >70”]; where track is constant as
Instrumentation and hometown Luzon.

```python
#(a)
Instru = eceboard.loc[(eceboard['Electronics'] > 70) &  #Locates all values in the Electronics column higher than 70.
                      (eceboard['Hometown'] == 'Luzon') & #Locates all values with 'Luzon' in the Hometown column 
                      (eceboard['Track'] == 'Instrumentation'), #Locates all values with 'Instrumentation' in the Track column.
                      ['Name', 'GEAS', 'Electronics']] #Displays only the columns Name, GEAS, and Electronics for those that fits all the conditions above.

Instru
```

<u>Name the file 'IMindy' and display the columns with the conditions [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is
constant as Mindanao and gender Female

```python
eceboard['Average'] = eceboard[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1) 
eceboard

#Adds a new column after the last column that computes the mean of the four subjects: Math, Electronics, GEAS, and Communication.
```

```python
Mindy = eceboard.loc[(eceboard['Average'] >= 55) & #Locates all values that are above or equals to 55 in the Average column. 
                      (eceboard['Hometown'] == 'Mindanao') & #Locates all values with 'Mindanao' in the Hometown column 
                      (eceboard['Gender'] == 'Female'), #Locates all values with 'Female' in the Gender column.
                      ['Name', 'Track', 'Electronics']] #Displays only the columns Name, Track, and Electronics for those that fits all the conditions above.

Mindy
```

2). Create a visualization that shows how the different features contributes to average grade.

```python
import matplotlib.pyplot as plt
```

```python
average1 = eceboard.groupby('Track')['Average'].mean()

plt.figure(figsize=(10, 10))
average1.plot(kind="bar")
plt.title('Track on Average Score')
plt.ylabel('Average Score')
plt.xlabel('Track')
plt.show()
```

```python
average2 = eceboard.groupby('Gender')['Average'].mean()

plt.figure(figsize=(10,10))
average2.plot(kind="bar")
plt.title('Gender on Average Score')
plt.ylabel('Average Score')
plt.xlabel('Gender')
plt.show()
```

```python
average3 = eceboard.groupby('Hometown')['Average'].mean()

plt.figure(figsize=(10, 10))
average3.plot(kind="bar")
plt.title('Hometown on Average Score')
plt.ylabel('Average Score')
plt.xlabel('Hometown')
plt.show()
```
